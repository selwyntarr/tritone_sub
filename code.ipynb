## Pivot Table

```python
import pandas as pd
import numpy as np
```

```python
#read xlsx file
data_df = pd.read_excel('skill_test_data.xlsx', sheet_name='data')
```

```python
data_df.head()
```

```python
data_df
```

```python
pivot = pd.pivot_table(data=data_df, 
                       values=['Spend','Attributed Rev (1d)','Imprs','Visits','New Visits','Transactions (1d)','Email Signups (1d)'], 
                       index=['Platform (Northbeam)'], 
                       aggfunc='sum', 
                       margins=True, 
                       margins_name='Grand Total'
                      )
pivoted_df = pivot.loc[:"YouTube Organic",['Spend','Attributed Rev (1d)','Imprs','Visits','New Visits','Transactions (1d)','Email Signups (1d)']].sort_values(by='Attributed Rev (1d)', ascending=False).append(pivot.loc['Grand Total'])
```

```python
pivoted_df
```

## SQLite3

```python
import sqlite3
connection = sqlite3.connect("test.db")
```

```python
pivoted_df.to_sql('pivot_table',connection,if_exists='replace',index=False)
connection.commit()
```

```python
cursor = connection.cursor()
cursor.execute("select * from pivot_table")
output = cursor.fetchall()
connection.commit()
connection.close()
output_df = pd.DataFrame(output)
```
